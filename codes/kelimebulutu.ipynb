```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Metni ve diğer parametreleri ayarlayın
text = "Cyberbullying ,Online harassment ,Cyberstalking,Doxing ,Trolling ,Flaming ,Outing,Catfishing ,Impersonation,Harassment ,Hate speech,Sexting,Onlineshaming,Exclusion ,Rumorspreading,blackmail,intimidation,Phishing,invasionofprivacy,manipulation identitytheft ,humiliation,Cyber-attacks,Onlineabuse,defamation,Cyber-victimization,Cyber-shaming,Online rumor-mongering  "
wordcloud = WordCloud(width=800, height=300).generate(text)

# Kelime bulutunu görselleştirin
plt.figure(figsize=(8, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()
```
